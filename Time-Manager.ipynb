```python
import time
import datetime
import os

# Function for getting inputs in hours,minutes,seconds
def countdown(h, m, s):
    
    try:
        pass

        try:
            # Calculating total hours based on given input
            Total_hours = h * 3600 + m * 60 + s
      
                
            # Getting inputs for morning status call
            morningstatuscallduration=int(input("Enter the morning status call duration in seconds"))
            morningstatuscall={"hours":input("Enter the morning status call hour"),"minutes":input("Enter the morning status minutes"),"seconds":input("Enter the morning status seconds")}
            M=morningstatuscall["hours"]
            N=morningstatuscall["minutes"]
            O=morningstatuscall["seconds"]
            
            # Getting inputs for morning break
            morningbreakduration=int(input("Enter the morning break duration in seconds"))
            morningbreak={"hours":input("Enter the morning break hour"),"minutes":input("Enter the morning break minutes"),"seconds":input("Enter the morning break seconds")}
            X=morningbreak["hours"]
            Y=morningbreak["minutes"]
            Z=morningbreak["seconds"]
            
            # Getting inputs for lunch break
            lunchbreakduration=int(input("Enter the lunch break duration in seconds"))
            lunchbreak={"hours":input("Enter the lunch break hour"),"minutes":input("Enter the lunch break minutes"),"seconds":input("Enter the lunch break seconds")}
            A=lunchbreak["hours"]
            B=lunchbreak["minutes"]
            C=lunchbreak["seconds"]
    
            # Gettig inputs for evening break
            eveningbreakduration=int(input("Enter the evening break duration in seconds"))
            eveningbreak={"hours":input("Enter the evening break hour"),"minutes":input("Enter the evening break minutes"),"seconds":input("Enter the evening break seconds")}
            D=eveningbreak["hours"]
            E=eveningbreak["minutes"]
            F=eveningbreak["seconds"]
        
            # Getting inputs for evening status call
            eveningmeetingduration=int(input("Enter the evening meeting duration in seconds"))
            eveningmeeting={"hours":input("Enter the evening meeting hour"),"minutes":input("Enter the evening meeting minutes"),"seconds":input("Enter the evening meeting seconds")}
            P=eveningmeeting["hours"]
            Q=eveningmeeting["minutes"]
            R=eveningmeeting["seconds"]
            
            # Getting inputs for logout remainder
            logoutremainder={"hours":input("Enter the logout remainder hour"),"minutes":input("Enter the logout remainder minutes"),"seconds":input("Enter the logout remainder seconds")}
            S=logoutremainder["hours"]
            T=logoutremainder["minutes"]
            U=logoutremainder["seconds"]
        
            # Calculating total working hours based on total hours and breaktime
            Total_working_hours = Total_hours - (morningbreakduration+lunchbreakduration+eveningbreakduration)
        
            # Calculating breakduration
            Total_break_duration = morningbreakduration+lunchbreakduration+eveningbreakduration
            
            # Printing total working hours
            print("Total_working_hours=",Total_working_hours)
            
            # Printing total break duration
            print("Total_break_duration=",Total_break_duration)
            
            # Calculating when the scheduled calls and break will occur
            MSC=Total_working_hours-((int(M)*3600)+(int(N)*60)+int(O))
            
            MB=Total_working_hours-((int(X)*3600)+(int(Y)*60)+int(Z))
                
            LB=Total_working_hours-((int(A)*3600)+(int(B)*60)+int(C))
                
            EB=Total_working_hours-((int(D)*3600)+(int(E)*60)+int(F))
            
            EM=Total_working_hours-((int(P)*3600)+(int(Q)*60)+int(R))
                
            LR=Total_working_hours-((int(S)*3600)+(int(T)*60)+int(U))
    
            
            try:
            # Function for timer using timedelta function
                while Total_working_hours > 0:
                    timer = datetime.timedelta(seconds = Total_working_hours)
                    print(timer, end="\r")
                    time.sleep(1)
                        
                    if Total_working_hours==MSC:
                        print("You have a status call for {} seconds".format(morningstatuscallduration))
                        break
                    Total_working_hours -= 1
            
                # Function for timer using timedelta function
                while Total_working_hours > 0:
                    timer = datetime.timedelta(seconds = Total_working_hours)
                    print(timer, end="\r")
                    time.sleep(1)
                        
                    if Total_working_hours==MB:
                        print("Take a morning tea break for {} seconds".format(morningbreakduration))
                        break
                    Total_working_hours -= 1
    
                # Function for breaktimer using divmod
                a=morningbreakduration
                while a: 
                    mins, secs = divmod(a, 60) 
                    timer = '{:02d}:{:02d}'.format(mins, secs) 
                    print(timer, end="\r") 
                    time.sleep(1) 
                    a -= 1
                print("Morning tea break is over,continue with your work")
        
                # Function for timer using timedelta function   
                while a==0:
                    timer = datetime.timedelta(seconds =Total_working_hours)
                    print(timer, end="\r")
                    time.sleep(1)
                        
                    if Total_working_hours==LB:
                        print("Take a lunch break for {} seconds".format(lunchbreakduration))
                        break
                    Total_working_hours -= 1
            
                # Function for breaktimer using divmod
                b=lunchbreakduration
                while b: 
                    mins, secs = divmod(b, 60) 
                    timer = '{:02d}:{:02d}'.format(mins, secs) 
                    print(timer, end="\r") 
                    time.sleep(1) 
                    b -= 1
                print("Lunch break is over,continue with your work")
        
                # Function for timer using timedelta function
                while b==0:
                    timer = datetime.timedelta(seconds =Total_working_hours)
                    print(timer, end="\r")
                    time.sleep(1)
                        
                    if Total_working_hours==EB:
                        print("Take a evening tea break for {} seconds".format(eveningbreakduration))
                        break
                    Total_working_hours -= 1
    
                # Function for breaktimer using divmod
                c=eveningbreakduration
                while c: 
                    mins, secs = divmod(c, 60) 
                    timer = '{:02d}:{:02d}'.format(mins, secs) 
                    print(timer, end="\r") 
                    time.sleep(1) 
                    c -= 1
                print("Evening tea break is over,continue with your work")
                
                # Function for timer using timedelta function
                while c==0:
                    timer = datetime.timedelta(seconds =Total_working_hours)
                    print(timer, end="\r")
                    time.sleep(1)
                    
                    if Total_working_hours==EM:
                        print("You have a evening meeting for {} seconds".format(eveningmeetingduration))
                        break
                    Total_working_hours -= 1
    
        
                # Function for timer using timedelta function
                while Total_working_hours>0:
                    timer = datetime.timedelta(seconds =Total_working_hours)
                    print(timer, end="\r")
                    time.sleep(1)
                        
                    if Total_working_hours==LR:
                        print("Your system will be autologged off after 5 seconds, save all your work immediately,enter your timesheets and update the daily reports to your RM")
                    Total_working_hours -= 1
                
                # Defining a function for shutdown
                def shutdown_computer():
                    if os.name == 'nt':  # For Windows
                        os.system('shutdown /s /t 1')
                    elif os.name == 'posix':  # For Linux/Unix/MacOS
                        os.system('sudo shutdown now')
                    else:
                         print("Unsupported operating system")
            
                # Calling shutdown function when timer completes
                if Total_working_hours==0:
                    print("Your working hours is completed")
                    print(shutdown_computer())
                    
            except:
                print("Invalid Time")
            
        finally:
            print("thank you")

    finally:
            print("thank you")    



# Inputs for hours, minutes, seconds on timer

try:
    # Get three inputs from the user
    h = input("Enter the time in hours: ")
    m = input("Enter the time in minutes: ")
    s = input("Enter the time in seconds: ")

    # Check if all inputs are zero
    if h == 0 and m == 0 and s == 0:
        raise ValueError("Time cannot be zero")


    print(countdown(int(h), int(m), int(s)))
    
except ValueError as ve:
    print("Error:", ve)
except Exception as e:
    print("An unexpected error occurred:", e)
```
